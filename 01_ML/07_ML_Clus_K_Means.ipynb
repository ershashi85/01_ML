***Clustering** is about creating **segmentation** of datasets (UNSUPERVISED) having similar characteristics. 

The main purpose of clustering can be for :
- Exploratory Data Analysis
- Summary generation or reducing the scale
- Outlier detection
- Fraud detection
- Noise removal
- Finding duplicates 
- Pre-processing step for either prediction, other data mining tasks or as a part of a complex system.

Types:
1. Partition Based Clustering
2. Hierarchical Based Clustering
3. Density Based Clustering

*1. Partitioned-Based Clustering*
is a group of clustering algorithms that produces sphere like clusters, such as k-means, or Fuzzy c-Means. Used for medium or large sized datasets.

*2. Heirarchical Based Clustering*
This algorithm produces trees of clusters, such as Agglomerative and Divisive algorithms. This group of algorithms are very intuitive and good for small size datasets.

*3. Density based clustering*
This algorithm produce arbitrary shaped clusters. They are good when dealing with spatial clusters or when there is noise in the dataset. eg.e DBSCAN algo.

1. Retail industry - to identify buying patterns.

3. Insurance industry - Fraud detection in claims
4. public Media - to auto categorize news based on its content.
5. Medicine - to identify patient behavior based on characterisitic.
etc...

**K-Means** algorithm is a type of partitioning clustering. It divides the data into k non-overlaping subsets(or clusters). The distance of samples from each other is used to shape the clusters. In general, the target is to minimize the "intra-cluster" distance and maximize the "inter-cluster" distance.

--The key concept of the k-Means algo is that it randomly picks a center point for each cluster. 

--Calculating each sample distance (Euclidean distance or many other) from the centroid to find the closest center (distance matrix showing each row as a sample distance).

-- Calculate total distance error.

--Shifting the centroids to minimize the distance error within each cluster. 

-- Distance calc and Centroids movement needs to in an iterative way until the algorithm converges (Centroid dont move further). 

-- Elbow method is a method to reduce distance error by increasing the k clusters. plotting the distance error with increase of k and finding out the right elbow point.



```python
import numpy as np
import random 
from sklearn.datasets.samples_generator import make_blobs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(0)
```

**k-Means ona randomly generated datasheet**

```python
X, y = make_blobs(n_samples = 5000, centers = [[4,4], [-2,-1], [2,-3], [1,1]], cluster_std = 0.9)
```

```python
plt.scatter(X[:, 0], X[:, 1], marker = '.')
```

**Setting up K-Means**

```python
from sklearn.cluster import KMeans
```

```python
k_means = KMeans(init = 'k-means++', n_clusters = 4, n_init = 12)
```

```python
k_means.fit(X)
```

```python
k_means_labels = k_means.labels_
```

```python
k_means_labels
```

```python
k_means_cluster_centers = k_means.cluster_centers_
k_means_cluster_centers
```
